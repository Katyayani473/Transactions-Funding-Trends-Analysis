```python
import pandas as pd
```

```python
df=pd.read_csv('startup_funding.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'Date dd/mm/yyyy':'Date'},inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
```

```python
df.drop('Remarks',axis=1,inplace=True)
```

```python
df.rename(columns={'City  Location':'Location'},inplace=True)
```

```python
df.columns
```

```python
df['Industry Vertical'] = df['Industry Vertical'].fillna("Unknown")
df['SubVertical'] = df['SubVertical'].fillna("Not Specified")
```

```python
df['Location']=df['Location'].fillna('Unknown')
```

```python
df['Investors Name']=df['Investors Name'].fillna('Undisclosed')
```

```python
df.rename(columns={'Amount in USD':'Amount'},inplace=True)
```

```python
df['IsAmountDisclosed'] = df['Amount'].apply(lambda x: 0 if pd.isna(x) else 1)
df['Amount'] = df['Amount'].fillna(0)
```

```python
df.rename(columns={'InvestmentnType':'InvestmentType'},inplace=True)
```

```python
df['InvestmentType'] = df['InvestmentType'].fillna("Unknown")
```

```python
df['Amount']=df['Amount'].str.replace(',','')
```

```python
df['Amount'] = df['Amount'].astype(str).str.replace('undisclosed', '0', regex=False)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce').fillna(0).astype(int)
```

```python
df['Industry Vertical'].value_counts()
```

```python
a=df.groupby('Industry Vertical')['Amount'].sum()
a.sort_values(ascending=False)
```

```python
b=df.groupby('Location')['Amount'].sum()
b.sort_values(ascending=False)
```

```python
df.replace('Bengaluru','Bangalore',inplace=True)
```

```python
df.replace('US','USA',inplace=True)
```
